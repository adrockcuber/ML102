## plt.plot()

```python
# plt.plot을 이용해서 그래프 그리기
x = np.arange(0, 6, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()

# x축을 지정하지 않고 y축을 지정하면 x축은 0부터 시작하는 수열이 됨
y = np.sin(x)
plt.plot(y)
plt.show()
```

```python
# xlabel, ylabel을 지정하면 그래프에 표시됨
plt.plot(x, y)
plt.xlabel("x축")
plt.ylabel("y축")
plt.show()
```

```python
# xlim, ylim을 지정하면 그래프에 표시됨
plt.plot(x, y)
plt.xlim(0, 6)
plt.ylim(-1, 1)
plt.show()
```

```python
# 그래프의 제목 표시
plt.plot(x, y)
plt.xlim(0, 6)
plt.ylim(-1, 1)
plt.title("sin(x)")
plt.xlabel("x축")
plt.ylabel("y축")
plt.show()
```

```python
# 그래프 범례 표시
plt.plot(x, y, label="sin(x)")
plt.xlim(0, 6)
plt.ylim(-1, 1)
plt.title("sin(x)")
plt.xlabel("x축")
plt.ylabel("y축")
plt.legend()
plt.show()
```

```python
# cim을 이용해서 그래프 그리기
x = np.arange(0, 6, 0.1)
y = np.sin(x)
plt.plot(x, y, c="r", linestyle="--", marker="o")
plt.show()
```

```python
# 전부 정리하면
import matplotlib.pyplot as plt

x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
plt.plot(x, y, "r--o", label="line1")
plt.title("Plotting x and y")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 4)
plt.ylim(0, 7)
plt.legend(loc="upper left")
plt.show()
```

## Extra Plots

```python
# Scatter Plot
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
plt.scatter(x, y, c="r", marker="o")
plt.show()
```

```python
# Pie Chart
labels = ["apple", "banana", "grape"]
sizes = [30, 20, 15]
plt.pie(sizes, labels=labels, autopct="%1.1f%%", shadow=True)
plt.title("Pie Chart")
plt.show()
```

```python
# Bar Chart
x = np.arange(0, 6, 0.1)
y = np.sin(x)
plt.bar(x, y)
plt.show()
```

```python
# Histogram
x = np.random.randn(1000)
plt.hist(x, bins=20)
plt.show()
```

```python
# 3D Plot

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap="coolwarm")
plt.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
